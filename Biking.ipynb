```python
url = 'https://raw.githubusercontent.com/Razzi86/Biking0010/main/datapbl_pgh.csv'
df = pd.read_csv(url)
# Dataset is now stored in a Pandas Dataframe
# Dataset shows when bike lanes were obstructed by vehicles
```

```python
df.head(1)
```

```python
# remove numbers from entries, just leaving the street names

df['address'] = df['address'].str.replace('\d+', '')
df['address'] = df['address'].str.replace(r', Pittsburgh,', '')
df['address'] = df['address'].str.replace(r',', '')
df['address'] = df['address'].str.replace(r'.', '')
df['address'] = df['address'].str.replace(r' Pittsburgh ', ' ')
df['address'] = df['address'].str.replace(r'PA ', 'PA')
df['address'] = df['address'].str.replace(r'PA', '')
df['address'] = df['address'].str.replace(r'th Ave', '')
df['address'] = df['address'].str.replace(r'th St', '')
df['address'] = df['address'].str.replace(r'US', '')

# df['address'] = df['address'].str.split().str[0]
streets = df['address']

streets.head(1)

street_counts = {}

for street in streets:
  if street not in street_counts:
    street_counts[street] = 1
  else:
    street_counts[street] += 1

street_counts = dict(sorted(street_counts.items(), key=lambda item: item[1]))

for k, v in street_counts.items():
    print(k + ": " + str(v))
```

```python
# this information points out streets with most obstructed bike paths
plt.bar(street_counts.keys(), street_counts.values())
plt.xticks(rotation=90)
plt.xlabel('Street Names')
plt.ylabel('# of obstructions')
plt.title('Number of bike lane obstructions per street')
plt.show()
```

2.) Brighton Heights​

3.) City of Pittsburgh​

4.) Strip District

```python
url = 'https://raw.githubusercontent.com/Razzi86/Biking0010/main/6dfd4f8f-cbf5-4917-a5eb-fd07f4403167.csv'
df_traffic = pd.read_csv(url)
# Dataset is now stored in a Pandas Dataframe
# Dataset shows number of speeders per neighborhood

df_traffic.head(10)
```

```python
neighborhoods = df_traffic['neighborhood']

neighborhood_counts = {}

for neighborhood in neighborhoods:
  if neighborhood not in neighborhood_counts:
    neighborhood_counts[neighborhood] = 1
  else:
    neighborhood_counts[neighborhood] += 1

neighborhood_counts = dict(sorted(neighborhood_counts.items(), key=lambda item: item[1]))

for k, v in neighborhood_counts.items():
    print(k , ": ",  str(v))
```

```python
plt.bar(range(len(neighborhood_counts)), list(neighborhood_counts.values()), align='center')
plt.xticks(range(len(neighborhood_counts)), list(neighborhood_counts.keys()), fontsize=4.7, rotation='vertical')
plt.xlabel('Neighborhood')
plt.ylabel('Frequency')
plt.title('Number of speeders per week per neighborhood')

plt.show()
```

```python
url = 'https://raw.githubusercontent.com/Razzi86/Biking0010/main/healthy-ride-station-locations-q4-2021.csv'
df_rent = pd.read_csv(url)
# Dataset is now stored in a Pandas Dataframe
# Dataset shows what streets in which neighborhood have the most bike stations
```

```python
df_rent = df_rent.sort_values("# of Racks")
df_rent.tail(10)
```

```python
stations = df_rent['Station Name']
df_rent['Station Name'] = df_rent['Station Name'].str.replace(r'.', '')
street1 = df_rent['Station Name'].str.split('&').str[0].str.strip()
street2 = df_rent['Station Name'].str.split('&').str[1].str.strip()

station_names = {}

for street in street1:
  if street not in station_names:
    station_names[street] = 1
  else:
    station_names[street] += 1

for street in street2:
  if street not in station_names:
    station_names[street] = 1
  else:
    station_names[street] += 1

station_names = dict(sorted(station_names.items(), key=lambda item: item[1]))

for k, v in station_names.items():
    print(k , ": " , str(v))
```

```python
plt.bar(range(len(station_names)), list(station_names.values()), align='center')
plt.xticks(range(len(station_names)), list(station_names.keys()), fontsize=4, rotation='vertical')
plt.xlabel('Street')
plt.ylabel('Biking Stations')
plt.title('Number of accessible bike racks per street')

plt.show()
```
